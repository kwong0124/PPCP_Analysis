```python
import pandas as pd
import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/SortByType.csv')
```

```python
data.columns
```

```python
cols = ['Author', 'Compound', 'Tech', 'RL_inf', 'Influent', 'RL_eff',
       'Effluent', 'Type']
data = data[cols]
data.head()
```

```python
data.dtypes
```

```python
data.iloc[530]
```

```python
data = data.drop(530)
```

```python
data['Influent'] = pd.to_numeric(data['Influent'])
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data['Log_inf'] = data['Influent'].apply(lambda x: np.log(x))
data['Log_eff'] = data['Effluent'].apply(lambda x: np.log(x))
```

```python
def removal(row):
    
    inf = row['Influent']
    eff = row['Effluent']
    
    return (inf - eff) / inf
```

```python
data['Removal'] = data.apply(removal, axis=1)
data['Log_removal'] = data['Removal'].apply(lambda x: np.log(1-x))
```
